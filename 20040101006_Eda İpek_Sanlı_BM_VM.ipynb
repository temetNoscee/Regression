<b>Veri Seti Hakkında Bilgiler</b><br>
Verileri Coffee Quality Institute'ten alan "Coffee Quality Data" veri seti, arabica türü kahve üretimi, işlenmesi ve duyusal analize ilişkin çeşitli veriler içermektedir. Aynı zamanda kahve genetiği, toprak türleri ve kahve kalitesini etkileyebilecek diğer faktörlere ilişkin verileri de içerir. CQI dünya çapında kahvenin kalitesini ve değerini artırmak için çalışan, kar amacı gütmeyen bir kuruluştur. 1996 yılında kurulmuştur ve merkezi ABD'nin Kaliforniya eyaletinde bulunmaktadır.
<b> Kahve Kalitesi Hakkındaki Değişkenlerin Bilgileri </b><br>
-Aroma: Kahvenin kokusunu veya kokusunu ifade eder.<br>
-Lezzet: Kahvenin tadı, tatlılık, acılık, asitlik ve diğer tat notları da dahil olmak üzere tada göre değerlendirilir.<br>
-Ağızda Kalan Tat: Kahveyi yuttuktan sonra ağızda kalan tadı ifade eder.<br>
-Asitlik: Kahvedeki asitlik, tadın parlaklığını veya canlılığını ifade eder.<br>
-Gövde: Kahvenin gövdesi, kahvenin ağızdaki kalınlığını veya viskozitesini ifade eder.<br>
-Denge: Denge, kahvenin farklı lezzet bileşenlerinin birlikte ne kadar iyi çalıştığını ifade eder.<br>
-Tekdüzelik: Tekdüzelik, kahvenin fincandan fincana tutarlılığını ifade eder.<br>
-Temiz Fincan: Temiz fincan, ekşilik, küflenme veya bayatlık gibi herhangi bir kötü tat veya kusur içermeyen kahveyi ifade eder.<br>
-Tatlılık: Karamel benzeri, meyvemsi veya çiçeksi olarak tanımlanabilir ve kahvede arzu edilen bir kalitedir.<br>

<b>Defects(Kusurlar)</b><br>
Kusurlar, kahve çekirdeklerinde işleme veya depolama sırasında oluşabilecek istenmeyen niteliklerdir. Kusurlar iki kategoriye ayrılabilir: Kategori bir ve kategori iki kusurlar.<br>
-Kategori Bir kusurlar, kahve çekirdeklerinin görsel olarak incelenmesiyle algılanabilen birincil kusurlardır. Bu kusurlar arasında Siyah fasulye, ekşi fasulye, böceklerden zarar görmüş fasulye, mantardan zarar görmüş fasulye vb. bulunur.<br>
-Kategori İki kusurlar, daha incelikli olan ve yalnızca tadım yoluyla tespit edilebilen ikincil kusurlardır. Bu kusurlar arasında aşırı fermantasyon, bayatlık, kokuşmuşluk, kimyasal tat vb. yer alır.

<b>Problemin Tanımı:</b><br>
Amacım, coffee quality veri setinde kahvenin özelliklerine göre "total cup points" değerlerini tahmin etmeye çalışan bir regresyon modeli oluşturmak.


-Sertifika İletişimi Toplam Kupa Puanı ile ilişkilendirilemez
Sertifikasyon Kuruluşunun Sertifika Adresi ülke adı olduğundan Sertifika Adresi atlanacak
Lot Numarasının hedef değişken üzerinde etkisi yoktur.
Menşe Ülkemiz olacağı için rakım düşecek
Çiftlik Adı ve Değirmen neredeyse aynı değere sahip olduğundan Değirmen düşürülecek.
Durumun yalnızca 1 değeri vardır, bu nedenle düşürülecektir.
Toplam Kupa Puanını Kullan. Aroma, Lezzet, Ağızda Kalan Tat, Asitlik, Gövde, Denge, Tekdüzelik, Temiz Bardak, Tatlılık ve Genel olarak düşme özelliği

```python
data2 = df[['Flavor','Balance','Overall','Total Cup Points']]
data2.info()
```

```python


correlation_matrix = data2[['Flavor','Balance','Overall','Total Cup Points']].corr()

# Heatmap çizimi
plt.figure(figsize=(8, 8))
sns.heatmap(correlation_matrix[['Total Cup Points']], annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5, cbar_kws={'location': 'right'})
plt.title('Total Cup Points ile Kahve Özellikleri Arasındaki Korelasyon')
plt.show()

data2.columns

```

```python
X = data2.drop('Total Cup Points', axis=1)
y = data2['Total Cup Points']

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression

results_list = []

for _ in range(10):
    # Veri setini böl (yüzde 80 train, yüzde 20 test)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
    model = LinearRegression()
    # Modeli eğit
    model.fit(X_train, y_train)
    
    # Test seti üzerinde tahmin yap
    y_pred = model.predict(X_test)

    # Performans metriklerini hesapla
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    # Sonuçları listeye ekle
    results_list.append({'Mean Absolute Error': mae,
                         'Mean Squared Error': mse,
                         'R^2 Score': r2})

# Listeyi bir DataFrame'e çevir
results_df = pd.DataFrame(results_list)

# Sonuçları yazdır
print(results_df)
```
